```python
from pymantic import sparql
```

```python
server = sparql.SPARQLServer('http://127.0.0.1:9999/bigdata/sparql')
```

```python
query_txt = """
SELECT distinct ?leftLabel ?prop1 ?shared1Label ?prop12 ?shared2Label ?shared1 ?shared2
WHERE
{
    values ?left {wd:Q319541}
    ?left ?prop1 ?shared1 .
    FILTER NOT EXISTS {?shared1 rdf:type ?type .}
    ?shared1 ?prop12 ?shared2 .
    FILTER NOT EXISTS {?shared2 rdf:type ?type .}
  ?shared2 ?prop2 ?right .	
    FILTER NOT EXISTS {?right rdf:type ?type .}
    SERVICE wikibase:label { bd:serviceParam wikibase:language "en" }
} limit 1000"""
```

```python
result = server.query(query_txt)
```
